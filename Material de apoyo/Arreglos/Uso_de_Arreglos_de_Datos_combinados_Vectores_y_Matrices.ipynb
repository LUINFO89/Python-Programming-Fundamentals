```python
import random

#Función para escribir la matriz
def escribirMatriz(matriz):
  for i in range(len(matriz)):
    for j in range(len(matriz[i])):
      print(matriz[i][j], end=' ')
    print()

#función para calcular el producto escalar
def escalar(matriz,valor):
  for i in range(len(matriz)):
    for j in range(len(matriz[i])):
      matriz[i][j] = matriz[i][j] * valor
  escribirMatriz(matriz) #llamado a la función para mostrar la matriz resultante
  
def main():  
  filas = int(input("digite el tamaño de las filas:"))
  columnas = int(input("digite el tamaño de las columnas:"))
  x = int(input("digite un número:"))
  
  matrizA = []
  #lleno la matriz
  for i in range(filas):
    fila = []
    for j in range(columnas):
      valor = random.randint(1,9)
      fila.append(valor)
    matrizA.append(fila)

  escribirMatriz(matrizA)   
  escalar(matrizA,x)  

main()
```
