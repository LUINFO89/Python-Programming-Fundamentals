```
def factorial(numero):
  fact = 1
  for i in range (1,numero+1):
    fact = fact * i
  print(numero, fact)
  return fact


def combinatoria(n,m):
  factN = factorial(n)#5
  factM = factorial(m)#3
  factNM = factorial(n-m)#2
  resultado = factN / (factM * factNM)
  print(resultado)

combinatoria(5,3)
```
